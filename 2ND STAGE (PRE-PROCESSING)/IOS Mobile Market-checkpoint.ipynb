```python
def sorgula(row):
    if(row["price"]>0):
        return 1;
    else:
        return 0;
```

```python
def aykirideger(row):
    if( row["sup_devices.num"]<=(36+11)/2):
        row["sup_devices.num"]=0
    else:
        row["sup_devices.num"]
```

```python
def isfree(row):
    if(row["price"]>0):
        row["free"]= row["price"]
    else:
        row["notFree"]= row["price"]
```
